```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_vendas_clientes.json')
dados.head()
```

```python
dados = pd.json_normalize(dados['dados_vendas'])
dados
```

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[1:])
dados
```

```python
dados.reset_index(inplace=True, drop=True)
dados
```

```python
dados.info()
```

```python
dados['Valor da compra'] = dados['Valor da compra'].apply(lambda x: x.replace('R','').replace('$','').replace(',','.').strip())
```

```python
import numpy as np
```

```python
dados['Valor da compra'] = dados['Valor da compra'].astype(np.float64)
```

```python
dados.info()
```

```python
dados['Cliente'] = dados['Cliente'].str.upper()
```

```python
dados['Cliente'] = dados['Cliente'].str.replace('[^a-zA-Z ]', '', regex=True).str.strip()
```

```python
dados
```

```python
dados['Data de venda'] = pd.to_datetime(dados['Data de venda'], format='%d/%m/%Y')
```

```python
dados.info()
```

```python
total_compras = dados.groupby(['Cliente'])['Valor da compra'].sum()
total_compras
```
